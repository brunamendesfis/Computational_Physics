# Exercício 2:

Repita o exercício 1 para a função U(x) = x²(x − 1)(x + 1). Esta função tem dois mínimos globais. Use x_0 = 2 e tente ajustar α para tentar fazer o código convergir ora num mínimo, ora no outro. O que acontece? O que você pode concluir sobre a escolha da taxa de aprendizado α?

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return x**2*(x - 1)*(x + 1)

def derivada(x):
    grad = 2*x**3 + 2*x**2 - 2*x
    return grad
```

```python
def grad_desc(x_0, alpha, itera, epsilon = 0.01):
    x = x_0
    x_list = [x]
    y_list = [funcao(x)]

    for _ in range(itera):
        x_1 = x_0 - alpha * derivada(x)
        delta_x = x_1 - x_0

        if abs(delta_x) < epsilon:
            break

        x = x_1
        x_list.append(x)
        y_list.append(funcao(x))

    return x_list, y_list

x_0 = 2
itera = 1000
alpha = 0.01
epsilon = 0.01

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
alpha = 0.05

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```
