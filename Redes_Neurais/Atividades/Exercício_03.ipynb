# Exercício 3:

Repita o exercício 2, mas agora vamos manipular a altura dos mínimos somando uma reta em U(x), tal que a função agora é U(x) = x²(x − 1)(x + 1) + x/4.

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return x**2*(x - 1)*(x + 1) + x/4

def derivada(x):
    return 2*x**3 + 2*x**2 - 2*x + x
```

```python
def grad_desc(x_0, alpha, itera, epsilon):
    x = x_0
    x_list = [x]
    y_list = [funcao(x)]

    for _ in range(itera):
        x_1 = x_0 - alpha*derivada(x)
        delta_x = x_1 - x_0

        if abs(delta_x) < epsilon:
            break

        x = x_1
        x_list.append(x)
        y_list.append(funcao(x))

    return x_list, y_list

x_0 = 2
itera = 1000
epsilon = 0.01
alpha = 0.01

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 3, 3, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
alpha = 0.15

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 3, 3, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```
