# Exercício 1:
Implemente o algoritmo de gradiente descendente para encontrar o mínimo da função:

$$U(x) = x^2− 1$$

É um caso bem simples para o qual sabemos a solução exata. Ilustre o algoritimo com um gráfico mostrando a função $U(x)$ e a trajetória da partícula. Use inicialmente uma taxa de aprendizado $\alpha = 0.1$ e uma tolerância $\epsilon = 0.01$. O número máximo de iterações deve ser $1000$. A posição inicial da partícula deve ser $x_{0} = 5$. Depois, varie estes parâmetros para ver como eles afetam a convergência do algoritmo.

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    y = x**2 - 1
    return y

def derivada(x):
    grad = 2*x
    return grad
```

```python
# Função grad_desc para calcular o Gradiente Descendente gerando um loop
def grad_desc(x_0, alpha, itera, epsilon = 0.01):
    x = x_0
    x_list = [x]
    y_list = [funcao(x)]

# Loop para guardar valores de x e y quantas vezes coloco para itera (iterações) 
    for _ in range(itera):
        x_1 = x - alpha * derivada(x)
        delta_x = x_1 - x_0

        # Se o valor de delta_x for menor que a tolerância indicada o código para o loop
        if abs(delta_x) < epsilon:
            break

        x = x_1
        x_list.append(x)
        y_list.append(funcao(x))
    
    return x_list, y_list

x_0 = 5
itera = 1000
epsilon = 0.01
alpha = 0.1

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-5.5, 5.5, 400)

plt.plot(x, funcao(x), label='y = x² - 1', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Mantendo $\alpha = 0.1$ e alterando $x_{0}$ para $2$, temos:

```python
x_0 = 2
alpha = 0.1

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-5, 5, 400)

plt.plot(x, funcao(x), label='y = x² - 1', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_0 = 5
alpha = 0.3

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-5.5, 5.5, 400)

plt.plot(x, funcao(x), label='y = x² - 1', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

No exemplo acima, assim como no primeiro, utilizei $x_{0} = 5$, mas aumentei o valor de $\alpha$ para $0.3$. Podemos verificar que os passos do Gradiente Descendente para atingir o mínimo da função diminui com o aumento da taxa de aprendizado.

```python
x_0 = 5
alpha = 0.5

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-5, 5, 400)

plt.plot(x, funcao(x), label='y = x² - 1', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Aumentando ainda mais o valor de $\alpha$ verificamos que o Gradiente Descendente atinge diretamente o mínimo da função. Isso sugere que o aumento da taxa de aprendizado indica a acurácia do gradiente descendente.
