# Exercício 2:

Repita o exercício 1 para a função:

$$U(x) = x^2(x − 1)(x + 1)$$

Esta função tem dois mínimos globais. Use $x_{0} = 2$ e tente ajustar α para tentar fazer o código convergir ora num mínimo, ora no outro.

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return x**2*(x - 1)*(x + 1)

def derivada(x):
    grad = 2*x**3 + 2*x**2 - 2*x
    return grad
```

```python
def grad_desc(x_0, alpha, itera, epsilon = 0.01):
    x = x_0
    x_list = [x]
    y_list = [funcao(x)]

    for _ in range(itera):
        x_1 = x_0 - alpha * derivada(x)
        delta_x = x_1 - x_0

        if abs(delta_x) < epsilon:
            break

        x = x_1
        x_list.append(x)
        y_list.append(funcao(x))

    return x_list, y_list

x_0 = 2
itera = 1000
alpha = 0.01
epsilon = 0.01

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Aumentando $\alpha$, temos:

```python
alpha = 0.05

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

É perceptível que o número de passos do Gradiente Descendente (GD) aumenta e tornam-se mais largos à medida que aumentamos o valor da taxa de aprendizado. Podemos visualizar também que com $\alpha$ maior, o GD tende para o mínimo da função. O que já era esperado, uma vez que realizei este procedimento para o exercício análogo anteriormente.

```python
alpha = 0.12

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Para um valor maior da taxa de aprendizado os passos de tormam ainda maiores. Podemos verificar que a partícula passou do mínimo da função para $\alpha = 0.12$, uma vez que visualizamos oscilações dos passos do GD entre os vales da função.

### O que acontece? O que você pode concluir sobre a escolha da taxa de aprendizado α?

Da mesma forma como no exercício 1, à medida que aumentamos a taxa de aperendizado o número de passos do Gradiente Descendente aumenta. Entretanto, como na função do exercício 2 temos dois mínimos, o valor de $\alpha$ terá um determinado gap que fará com que o GD atinja o mínimo da função e, acima deste gap como vimos para taxa de aprendizado acima de valores em torno de $0.12$, causará oscilações nos passos do GD.
