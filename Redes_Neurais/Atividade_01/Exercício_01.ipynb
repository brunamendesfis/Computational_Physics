# Exercise 1:
Implement the gradient descent algorithm to find the minimum of the function:

$$U(x) = x^2− 1$$

This is a very simple case for which we know the exact solution. Illustrate the algorithm with a plot showing the function $U(x)$ and the trajectory of the particle. Initially, use a learning rate $\alpha = 0.1$ and a tolerance $\epsilon = 0.01$. The maximum number of iterations should be $1000$. The initial position of the particle should be $x_{0} = 5$. Then, vary these parameters to see how they affect the convergence of the algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    y = x**2 - 1
    return y

def derivada(x):
    grad = 2*x
    return grad
```

```python
# grad_desc function to compute Gradient Descent using a loop
def grad_desc(x_0, alpha, itera, epsilon = 0.01):
    x = x_0
    x_list = [x]
    y_list = [funcao(x)]

# Loop to store values of x and y for each iteration (number of iterations)
    for _ in range(itera):
        x_1 = x - alpha * derivada(x)
        delta_x = x_1 - x_0

        # If the value of delta_x is less than the indicated tolerance, the code stops the loop
        if abs(delta_x) < epsilon:
            break

        x = x_1
        x_list.append(x)
        y_list.append(funcao(x))
    
    return x_list, y_list

x_0 = 5
itera = 1000
epsilon = 0.01
alpha = 0.1

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-5.5, 5.5, 400)

plt.plot(x, funcao(x), label='y = x² - 1', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Keeping $\alpha = 0.1$ and changing $x_{0}$ to $2$, we have:

```python
x_0 = 2
alpha = 0.1

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-5, 5, 400)

plt.plot(x, funcao(x), label='y = x² - 1', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_0 = 5
alpha = 0.3

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-5.5, 5.5, 400)

plt.plot(x, funcao(x), label='y = x² - 1', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

In the example above, as in the first one, I used $x_{0} = 5$, but increased the value of $\alpha$ to $0.3$. We can see that the steps of Gradient Descent to reach the minimum of the function decrease as the learning rate increases.

```python
x_0 = 5
alpha = 0.5

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-5, 5, 400)

plt.plot(x, funcao(x), label='y = x² - 1', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Increasing the value of $\alpha$ even further, we see that Gradient Descent reaches the minimum of the function directly. This suggests that the chosen increase in the learning rate determines the accuracy of gradient descent for this specific case.
