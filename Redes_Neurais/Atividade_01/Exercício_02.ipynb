# Exercise 2:

Repeat exercise 1 for the function:

$$U(x) = x^2(x − 1)(x + 1)$$

This function has two global minima. Use $x_{0} = 2$ and try adjusting α to make the code converge to one minimum or the other.

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return x**2*(x - 1)*(x + 1)

def derivada(x):
    grad = 4*x**3 - 2*x
    return grad
```

```python
def grad_desc(x_0, alpha, itera, epsilon = 0.01):
    x = x_0
    x_list = [x]
    y_list = [funcao(x)]

    for _ in range(itera):
        x_1 = x_0 - alpha * derivada(x)
        delta_x = x_1 - x_0

        if abs(delta_x) < epsilon:
            break

        x = x_1
        x_list.append(x)
        y_list.append(funcao(x))

    return x_list, y_list
```

```python
x_0 = 2
itera = 1000
alpha = 0.01
epsilon = 0.01

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Increasing $\alpha$:

```python
alpha = 0.04

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

It is noticeable that the number of steps in Gradient Descent (GD) increases and becomes wider as we increase the learning rate. We can also see that with a higher $\alpha$, GD tends toward a global minimum of the function.

```python
alpha = 0.06

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

For a higher learning rate, the steps become even larger. We can see that the particle has passed over one of the global minima of the function with $\alpha = 0.06$.

```python
alpha = 0.097

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(-2.5, 2.5, 400)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1)', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

We can verify the tendency of GD towards the second minimum of the function with $\alpha = 0.097$, however, the step becomes even wider. It is worth noting that this is the maximum value of the learning rate for this function using the parameters above.

### What happens? What can you conclude about the choice of the learning rate α?

As in exercise 1, as we increase the learning rate, the number of Gradient Descent steps increases. However, since the function in exercise 2 has two minima, the value of $\alpha$ will determine which minimum GD will tend towards, with the maximum value that $\alpha$ can take with these parameters being $0.097$.
