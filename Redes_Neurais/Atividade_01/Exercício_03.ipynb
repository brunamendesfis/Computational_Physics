# Exercício 3:

Repita o exercício 2, mas agora vamos manipular a altura dos mínimos somando uma reta em $U(x)$, tal que a função agora é 

$$U(x) = x²(x − 1)(x + 1) + x/4$$

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return x**2*(x - 1)*(x + 1) + x/4

def derivada(x):
    return 4*x**3 - 2*x + 1/4
```

```python
def grad_desc(x_0, alpha, itera, epsilon):
    x = x_0
    x_list = [x]
    y_list = [funcao(x)]

    for _ in range(itera):
        x_1 = x_0 - alpha*derivada(x)
        delta_x = x_1 - x_0

        if abs(delta_x) < epsilon:
            break

        x = x_1
        x_list.append(x)
        y_list.append(funcao(x))

    return x_list, y_list
```

```python
x_0 = 2
itera = 1000
epsilon = 0.01
alpha = 0.01

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
alpha = 0.04

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
alpha = 0.06

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

A partir de $\alpha = 0.06$ podemos visuaizar ligeira diferença nos passos do GD desta função em relação a função do exercício 2, onde a oscilação entre os passos diminui moderadamente. 

```python
alpha = 0.097

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Diferentemente da função do exercício 2, com $\alpha = 0.097$ foi necessário que o GD usasse mais passos para atingir a mesma região do gráfico. O que indica uma convergência mais lenta do GD neste caso.

```python
alpha = 0.098

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='Passos do GD')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradiente Descendente (GD) \n com α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

### O que acontece? O que você pode concluir sobre a escolha da taxa de aprendizado α?

Verificamos que para a função $U(x) = x^2(x - 1)(x + 1) + x/4$ é possível atingirmos uma taxa de aprendizado de $0.098$ (diferentemente da função do exercício 2 que atingia valores até $0.097$), sendo que o termo linear $x/4$ que diferencia as duas funções é responsável por suavizar a curva o que permite uma taxa de aprendizado maior para a função.
