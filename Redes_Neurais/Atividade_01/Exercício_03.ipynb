# Exercise 3:

Repeat exercise 2, but now let's manipulate the height of the minima by adding a straight line to $U(x)$, so that the function is now

$$U(x) = x^2(x - 1)(x + 1) + x/4$$

```python
import numpy as np
import matplotlib.pyplot as plt

def funcao(x):
    return x**2*(x - 1)*(x + 1) + x/4

def derivada(x):
    return 4*x**3 - 2*x + 1/4
```

```python
def grad_desc(x_0, alpha, itera, epsilon):
    x = x_0
    x_list = [x]
    y_list = [funcao(x)]

    for _ in range(itera):
        x_1 = x_0 - alpha*derivada(x)
        delta_x = x_1 - x_0

        if abs(delta_x) < epsilon:
            break

        x = x_1
        x_list.append(x)
        y_list.append(funcao(x))

    return x_list, y_list
```

```python
x_0 = 2
itera = 1000
epsilon = 0.01
alpha = 0.01

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
alpha = 0.04

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
alpha = 0.06

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

From $\alpha = 0.06$ onwards, we can observe a slight difference in the GD steps for this function compared to the function from exercise 2, where the oscillation between the steps decreases moderately.

```python
alpha = 0.097

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

Unlike the function from exercise 2, with $\alpha = 0.097$ it was necessary for GD to use more steps to reach the same region of the graph. This indicates a slower convergence of GD in this case.

```python
alpha = 0.098

xs, ys = grad_desc(x_0, alpha, itera, epsilon)

x = np.linspace(- 2.5, 2.5, 100)

plt.plot(x, funcao(x), label='U(x) = x²(x − 1)(x + 1) + x/4', color='black')
plt.scatter(xs, ys, color='red', label='GD Steps')
plt.plot(xs, ys, 'r--', alpha=0.5)
plt.title(f"Gradient Descent (GD) \n with α = {alpha}, ϵ = {epsilon}")
plt.legend()
plt.grid(True)
plt.show()
```

### What happens? What can you conclude about the choice of the learning rate α?

We found that for the function $U(x) = x^2(x - 1)(x + 1) + x/4$ it is possible to reach a learning rate of $0.098$ (unlike the function from exercise 2, which reached values up to $0.097$). The linear term $x/4$, which differentiates the two functions, is responsible for smoothing the curve, allowing a higher learning rate for this function.
